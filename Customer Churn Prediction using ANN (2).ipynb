```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("gauravtopre/bank-customer-churn-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
data=pd.read_csv('/kaggle/input/bank-customer-churn-dataset/Bank Customer Churn Prediction.csv')
data
```

```python
data.info()
```

```python
data.drop('customer_id', axis=1)
```

```python
non_numeric=data.select_dtypes('object')
non_numeric
```

```python
encoded=pd.get_dummies(non_numeric, drop_first=True, dtype=int)
encoded
```

```python
data = pd.concat([data.drop(columns=non_numeric),encoded],axis=1)
data
```

```python
data = data.drop('customer_id',axis=1)
data
```

```python
x = data.drop('churn',axis=1)
y = data['churn']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
len(x_train.columns)
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()

model.add(Dense(64, activation='relu', input_dim=11))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=15, validation_data=(x_test, y_test))
```

```python
yp = model.predict(x_test)
```

```python
yp[:10]
```

```python
ypred = []
for i in yp:
  if i > 0.5:
    ypred.append(1)
  else:
    ypred.append(0)
```
